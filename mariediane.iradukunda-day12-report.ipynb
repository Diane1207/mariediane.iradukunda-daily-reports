### Exercise 1:
Create a simple line plot for y = x for x values ranging from 0 to 10.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
```

```python
x = np.arange(1,10,1)
y = x

# create a figure/canvas of desired size
plt.figure(figsize=(8, 6))

# plot values; with a color `blue`
plt.plot(x, y, 'b')

# give labels to the axes
plt.xlabel('x')
plt.ylabel('y')

# give a title to the plot
plt.title(r"Plot of $y=x$")
plt.show()
```

### Exercise 4:
Plot a sine wave using Matplotlib for x values between 0 and 2π.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 256 points between 0 and 2*pi
X = np.linspace(0, 2 * np.pi, 256, endpoint=True)

S = np.sin(X) # Compute sine values

plt.plot(X, S) # Plot the sine curve

# Show the plot
plt.show()
```

##### Exercise 9:
Create a subplot with 2 rows and 1 column. Plot y1 = x^2 in the first subplot and y2 = x^3 in the second.

```python

# Generate values for x
x = np.linspace(0, 10, 100)

# Compute y1 and y2
y1 = x**2
y2 = x**3

# Create subplots: 2 rows, 1 column
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(6, 8))

# First subplot: y1 = x^2
ax1.plot(x, y1, color='blue', label='y = x^2')
ax1.set_title('plot of y = x^2')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.legend()

# Second subplot: y2 = x^3
ax2.plot(x, y2, color='green', label='y = x^3')
ax2.set_title('plot of y = x^3')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.legend()

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

#### Exercise 10:
Use a different color and linestyle for each line in a plot of three functions: y1 = x, y2 = x^2, and y3 = x^3.

```python

# Generate values for x
x = np.linspace(0, 10, 100)

# Compute y1, y2, and y3
y1 = x
y2 = x**2
y3 = x**3

# Create the plot
plt.figure(figsize=(8, 6))

# Plot y1 = x with a solid line
plt.plot(x, y1, color='blue', linestyle='-', label='y = x')

# Plot y2 = x^2 with a dashed line
plt.plot(x, y2, color='orange', linestyle='--', label='y = x^2')

# Plot y3 = x^3 with a dotted line
plt.plot(x, y3, color='green', linestyle=':', label='y = x^3')

# Add title and labels
plt.title("Plot of y = x, y = x^2, and y = x^3")
plt.xlabel("x")
plt.ylabel("y")

# Show the legend
plt.legend()

# Show the plot
plt.grid()
plt.show()
```

##### Exercise 11:
Create a pie chart with 5 segments labeled A, B, C, D, and E, with values [25, 35, 20, 15, 5].

```python

labels = ['A', 'B', 'C', 'D', 'E']
values = [25, 35, 20, 15, 5]

# Create pie chart
plt.figure(figsize=(8, 6))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90)

# Add title
plt.title("Pie Chart of Segments ")

# Equal aspect ratio ensures the pie chart is circular
plt.axis('equal')

# Show the plot
plt.show()
```

#### Exercise 14:
Plot a filled area under the curve of y = x^2 for x ranging from 0 to 10.

```python

x = np.linspace(0, 10, 100)
y = x**2  # y = x^2

# Create the plot
plt.figure(figsize=(8, 6))

# Plot the line
plt.plot(x, y, label='y = x^2', color='blue')

# Fill the area under the curve
plt.fill_between(x, y, color='skyblue', alpha=0.4)

# Add title and labels
plt.title("plot of y = x^2")
plt.xlabel("x")
plt.ylabel("y")

# Show the legend
plt.legend()

# Show the plot
plt.show()
```
